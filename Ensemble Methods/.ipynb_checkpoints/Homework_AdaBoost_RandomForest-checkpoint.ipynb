```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = [10, 5]
import timeit
from scipy import stats
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostRegressor, RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error, explained_variance_score, max_error, mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
#Data extraction and visualization
data = pd.read_csv("house_data.csv")
data.head(5)
```

```python
#Getting detials about the dataset
def data_exploration(df):   
    print('\n Dimensions of dataset is {}' .format(df.shape))
    print("\n Dataset is having {} rows and {} features \n" .format(df.shape[0],df.shape[1]))
    print('\n Datatype of features:\n {}' .format(df.dtypes))
    plt.figure(figsize=(15,10))
    df.dtypes.value_counts().plot(kind='bar')
    plt.title('Number of datatypes')
    plt.xlabel('Datatypes')
    plt.ylabel('Count')
    plt.grid()
    plt.show()
    print(df.dtypes.value_counts())
    print('\n Null data values in number of columns is {}' .format((df.isnull().any().sum())))
    print('\n Null data values in number of rows is {}'  .format((df.isnull().any(axis=1).sum())))

def unique_values(df):
    for col in df.columns:
        print("The unique values for the {} is \n {}"\
             .format(col, pd.unique(df[col])))
            
#Getting data correlation    
def data_correlations(feature_name, target_name, df):
    correlation = pd.DataFrame(columns=['correlation', 'value'])
    for feature in feature_name:
        feature_value = df[feature].values
        target_value = df[target_name].values
        correlation_value = stats.pearsonr(feature_value, target_value)[0]
        correlation = correlation.append({'correlation': feature+" & "+target_name,\
                                          'value': correlation_value}, ignore_index=True)
    correlation = correlation.sort_values(by='value', ascending=False)
    return correlation

#Visualization for data correlation
def visualization(correlation, df):
    best_correlation = correlation.index[:3]
    print("Will visualize the first {} features which has high correlation \n".format(len(best_correlation)))
    for features in best_correlation:
        feature_name = df.columns[3+features]
        target_name = df.iloc[:, 2].name
        plt.figure(figsize=(15,10))
        plt.scatter(df[feature_name],df[target_name])
        plt.xlabel(feature_name)
        plt.ylabel(target_name)
        plt.title("Plot of "+feature_name+" vs "+target_name)
        plt.grid()
        plt.show()

#One function to call the above function
def pre_processing(df):
    plt.rcParams["figure.figsize"] = [10, 5]
    data_exploration(df)
    feature_name = df.iloc[:, 3:].columns
    #unique_values(features)
    target_name = df.iloc[:, 2].name
    plt.figure(figsize=(15,10))
    correlation = data_correlations(feature_name, target_name, df)
    correlation.plot(x = "correlation", y= "value", kind='bar')
    plt.title('Correlation between features and price')
    plt.xlabel('Features vs price')
    plt.ylabel('correlation value')
    plt.grid()
    plt.show()
    visualization(correlation, df)
```

```python
#Preprocessing function call
pre_processing(data)
```

```python
#Function to plot the complexity curve
def Complexity_curve(estimatorType,tree_number,Xtrain,Ytrain,Xtest, Ytest):
    tree_range = np.arange(1,tree_number)  
    trainscore = np.zeros(len(tree_range)) 
    testscore = np.zeros(len(tree_range))  
    
    for i in range(len(tree_range)):
        if estimatorType == 'AdaBoostRegressor':            
            estimator = AdaBoostRegressor(n_estimators = tree_range[i])
            estimator.fit(Xtrain,Ytrain)
            trainscore[i] = explained_variance_score(Ytrain,estimator.predict(Xtrain)) 
            testscore[i] = explained_variance_score(Ytest,estimator.predict(Xtest))    
        
        elif estimatorType == 'RandomForestRegressor':
            estimator = RandomForestRegressor(n_estimators = tree_range[i])
            estimator.fit(Xtrain,Ytrain)
            trainscore[i] = explained_variance_score(Ytrain,estimator.predict(Xtrain)) 
            testscore[i] = explained_variance_score(Ytest,estimator.predict(Xtest))    
            
    # plot the model complexity curves i.e the training error and test error
    plt.figure(figsize=(15,10))
    plt.title('Complexity curves for %s' % (estimatorType))
    plt.plot(tree_range,testscore,lw = 2,label = 'Test score')
    plt.plot(tree_range,trainscore,lw = 2, label = 'Train score')
    plt.xlabel('Tree Number')
    plt.ylabel('Variance Score')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
#Function for AdaBoost Regressor
def adaboost(x_train, x_test, y_train, y_test):
    start_time = timeit.default_timer()
    adaboost = AdaBoostRegressor(n_estimators=100, learning_rate=1, loss="exponential")
    adaboost.fit(x_train, y_train)
    prediction = adaboost.predict(x_test)
    run_time = timeit.default_timer()-start_time
    return prediction, run_time

#Function for Random Forest Regressor
def random_forest(x_train, x_test, y_train, y_test):
    start_time = timeit.default_timer()
    random_forest = RandomForestRegressor(n_estimators=100)
    random_forest.fit(x_train, y_train)
    prediction = random_forest.predict(x_test)
    run_time = timeit.default_timer()-start_time
    return prediction, run_time
    
```

```python
features = data.drop(columns=["id", "date", "price", "zipcode"], axis=1)
target = data["price"]
#pre-processing of data
features = (features-features.mean(axis=0))/features.std(axis=0)
target = (target-target.mean(axis=0))/target.std(axis=0)


x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.25,random_state=0)

adaboost_pred, adaboost_time = adaboost(x_train, x_test, y_train, y_test)
exp_variance_score_adaboost = explained_variance_score(y_test, adaboost_pred)
print("\n The variance score for AdaBoost is {:.4f}\n".format(exp_variance_score_adaboost))
print("\n The R2 score for AdaBoost is {:.4f}\n".format(r2_score(y_test,adaboost_pred)))
# mse_adaboost = mean_squared_error(y_test,adaboost_pred)
# print("\nThe Mean Square Error for Adaboost is {:.4f}\n".format(mse_adaboost))
print("\n Time taken by Adaboost is {:.2f} seconds\n".format(adaboost_time))
max_error_adaboost = max_error(y_test, adaboost_pred)
mean_absolute_error_adaboost = mean_absolute_error(y_test, adaboost_pred)
print("\n The max error for AdaBoost is {:.4f}\n".format(max_error_adaboost))
print("\n The mean absolute error for AdaBoost is {:.4f}\n".format(mean_absolute_error_adaboost))

random_forest_pred, rf_time = random_forest(x_train, x_test, y_train, y_test)
exp_variance_score_rf = explained_variance_score(y_test, random_forest_pred)
print("\n The variance score for Random Forest is {:.4f}\n".format(exp_variance_score_rf))
print("\n The R2 score for Random forest is {:.4f}\n".format(r2_score(y_test, random_forest_pred)))
# mse_rf = mean_squared_error(y_test,random_forest_pred)
# print("\nThe Mean Square Error for Random Forest is {:.4f}\n".format(mse_rf))
print("\n Time taken by Random Forest is {:.2f} seconds\n".format(rf_time))
max_error_rf = max_error(y_test, random_forest_pred)
mean_absolute_error_rf = mean_absolute_error(y_test, random_forest_pred)
print("\n The max error for Random Forest is {:.4f}\n".format(max_error_rf))
print("\n The mean absolute error for Random Forest is {:.4f}\n".format(mean_absolute_error_rf))
   
```

```python
#Comparison between two methods
if exp_variance_score_rf > exp_variance_score_adaboost:
    print("\n The variance score of Random Forest is higher which signifies it perform best comparative to AdaBoost \n")
else:
    print("\n The variance score of AdaBoost is higher which signifies it perform best comparative to random Forest \n")

if adaboost_time > rf_time:
    print("\n The time taken by AdaBoost is more than the Random Forest  \n")
else:
    print("\n The time taken by Random Forest is more than the AdaBoost \n")
```

```python
#Plotting the complexity curve for 20 trees for each regression

estimatorType = 'AdaBoostRegressor'
Complexity_curve(estimatorType,20,x_train,y_train,x_test,y_test) 
estimatorType = 'RandomForestRegressor'
Complexity_curve(estimatorType,20,x_train,y_train,x_test,y_test) 
```

```python
#Executing this will take atleast 10mins
estimatorType = 'AdaBoostRegressor'
Complexity_curve(estimatorType,100,x_train,y_train,x_test,y_test) 
estimatorType = 'RandomForestRegressor'
Complexity_curve(estimatorType,100,x_train,y_train,x_test,y_test) 
```
